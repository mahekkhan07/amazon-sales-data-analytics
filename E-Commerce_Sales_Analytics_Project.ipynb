### Step-2 : Data Cleaning Process
- Removing Irrelevant columns.

```python
df = df.drop(columns=['Unnamed: 22'])
df.head()
```

### Step-3 : Check Missing / Null Values

```python
df.isnull().sum()
```

**Observed Missing Columns:-**

- Courier Status
- currency
- Amount
- ship-city               
- ship-state               
- ship-postal-code         
- ship-country             
- promotion-ids
- fulfilled-by

### Step - 4 : Missing Data Treatment

```python
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

- Convert Amount to numeric & fill missing values

```python
df['promotion-ids'] = df['promotion-ids'].fillna('No Promotion')
df['currency'] = df['currency'].fillna('INR')
df['ship-postal-code'] = df['ship-postal-code'].fillna('Unknown')
df['ship-country'] = df['ship-country'].fillna('India')
df['Courier Status'] = df['Courier Status'].fillna(df['Courier Status'].mode()[0])
df['ship-city'] = df['ship-city'].fillna(df['ship-city'].mode()[0])
df['ship-state'] = df['ship-state'].fillna(df['ship-state'].mode()[0])
df['fulfilled-by'] = df['fulfilled-by'].fillna(df['fulfilled-by'].mode()[0])
```

- Fill missing values in Categorical Columns

```python
df.isnull().sum()
```

- Recheck Missing Values

### Step - 5 : Convert Date Column into Date Format

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Date'].dtype
```

### Step 6 : Exploratory Data Analysis (EDA)¶

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Q1 — What is the total revenue generated from all orders?

```python
total_revenue = df['Amount'].sum()
print("Total Revenue:", total_revenue)
```

### Analysis :
- I added all values in the Amount column.
### Conclusion :
- Total sales revenue is very high, showing strong product sales.

### Q2 — Which order status occurs the most?

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Status', data=df)
plt.xticks(rotation=45, ha='right')
plt.title("Order Status Count")
plt.tight_layout()
plt.savefig("Order_Status_Count.png")
plt.show()
```

### Analysis :
- I counted how many orders fall under each status and plotted a bar chart.
### Conclusion :
- Most orders are delivered or shipped successfully.

### Q3 — Which states generate the highest sales?

```python
state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
state_sales.plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.ylabel("Revenue")
plt.savefig("state_sales.png")
plt.show()
```

### Analysis :
-  I summed the sales Amount state-wise.
### Conclusion :
- Few major states contribute the highest revenue.

### Q4 — Which cities contribute the highest revenue?

```python
state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
state_sales.plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.ylabel("Revenue")
plt.savefig("city_sales.png")
plt.show()
```

### Analysis :
-  I calculated total sales for each city.
### Conclusion :
- Metro cities generate the most revenue.

### Q5 — What is the order trend over time?

```python
daily_orders = df.groupby('Date')['Order ID'].count()

plt.figure(figsize=(10,5))
daily_orders.plot()
plt.title("Order Trend Over Time")
plt.ylabel("Number of Orders")
plt.savefig("order_trend.png")
plt.show()
```

### Analysis :
-  I grouped orders by Date and plotted order counts.
### Conclusion :
- Orders fluctuate daily with some busy days.

### Q6 — Which product category generates the highest sales?

```python
category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title("Top Categories by Revenue")
plt.ylabel("Revenue")
plt.savefig("category_sales.png")
plt.show()
```

### Analysis :
- I summed sales for each category.
### Conclusion :
- A few product categories perform much better than others.

### Q7 — Which fulfillment channel performs better? 

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Fulfilment', data=df)
plt.title("Fulfillment Type Count")
plt.savefig("fulfillment_type_count.png")
plt.show()
```

### Analysis :
- I compared sales between Amazon-fulfilled and Merchant-fulfilled orders.
### Conclusion :
-  Amazon-fulfilled orders perform better overall.

### Q8 — What percentage of orders are B2B vs retail?

```python
b2b_sales = df.groupby('B2B')['Amount'].sum()

plt.figure(figsize=(6,4))
b2b_sales.plot(kind='bar')
plt.title("B2B vs Retail Sales")
plt.ylabel("Revenue")
plt.savefig("b2b_vs_retail.png")
plt.show()
```

### Analysis :
-  I separated B2B and retail orders and compared sales.
### Conclusion :
- Retail (B2C) orders are much higher than B2B.

### Q9 — Which shipping service level is used the most?

```python
plt.figure(figsize=(8,5))
sns.countplot(x='ship-service-level', data=df)
plt.xticks(rotation=45, ha='right')
plt.title("Shipping Service Level Usage")
plt.tight_layout()
plt.savefig("shipping_service_level.png")
plt.show()
```

### Analysis :
- I counted the frequency of each shipping level.

### Conclusion :
- Standard shipping is used the most.

### Q10 — What is the highest order amount?

```python
max_order = df['Amount'].max()
print("Highest Order Amount:", max_order)
```

### Analysis :
